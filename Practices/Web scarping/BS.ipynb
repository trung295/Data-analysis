```python
Write a Python program to extract all the URLs from the webpage python.org that are nested within <li> tags from .
```

```python
from bs4 import BeautifulSoup
import requests

url = "https://www.python.org"

respond = requests.get(url)
print(f"Content type of website is {respond.headers["Content-Type"]}")

#parse the HTML
soup = BeautifulSoup(respond.text,"html.parser")
# find all li tag
lis = soup.find_all("li")
href =[]
valid_url =[]
for li in lis:
    a_tags = li.find_all("a")
    for a_tag in a_tags:
        href.append(a_tag.get("href"))
for full_url in href:
    if full_url.startswith("http"):
        valid_url.append(full_url)
print(valid_url)
    
        
```

```python
Write a Python program to a list of all the h1, h2, h3 tags from the webpage python.org.
```

```python
from bs4 import BeautifulSoup
import requests

url = "https://www.python.org"

respond = requests.get(url)

soup = BeautifulSoup(respond.text,"html.parser")
h_all =[]
#find all h1
h1_tags = soup.find_all("h1")
for h1_tag in h1_tags:
    h_all.append(h1_tag)
#find all h2
h2_tags = soup.find_all("h2")
for h2_tag in h2_tags:
    h_all.append(h2_tag)
#find h3 tags
h3_tags = soup.find_all("h3")
for h3_tag in h3_tags:
    h_all.append(h3_tag)
for heading in h_all:
    print(heading.name +" "+ heading.text)
```

```python
Write a Python program to retrieve children of the html tag from a given web page.
```

```python
from bs4 import BeautifulSoup
import requests

url = "https://www.python.org/"

respond = requests.get(url)

# parse the html
soup = BeautifulSoup(respond.text,"html.parser")

html_tag = soup.find("html")
for child in html_tag.children:
    if child.name is not None:
        print(child.name)
```

```python
Write a Python program to retrieve all descendants of the body tag from a given web page.
```

```python
from bs4 import BeautifulSoup
import requests

url = "https://www.python.org/"

respond = requests.get(url)

# parse the html
soup = BeautifulSoup(respond.text,"html.parser")

body = soup.find("body")
for des in body.descendants:
    if des.name is not None:
        print(des.name)
```
