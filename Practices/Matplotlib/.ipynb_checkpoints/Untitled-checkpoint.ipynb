```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([[0,6],[1,2]])
ypoints = np.array([0,250])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.arange(-10,11)
ypoints = xpoints**2

plt.plot(xpoints,ypoints,marker='o', linestyle='-', color='b', label="y = x")
plt.title("Exercises 1")
plt.xlabel("axis")
plt.ylabel("axis")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints =np.arange(1,10)
y1 = xpoints
y2 = xpoints**2
y3 = xpoints**3
plt.plot(xpoints,y1,marker='o', linestyle='-', color='r', label="y = x")
plt.plot(xpoints,y2,marker='o', linestyle='-', color='b', label="y = x")
plt.plot(xpoints,y3,marker='o', linestyle='-', color='y', label="y = x3")
plt.grid(True)
plt.legend()
plt.show()
```
