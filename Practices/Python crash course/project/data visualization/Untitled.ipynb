```python
import matplotlib.pyplot as plt
from matplotlib import style
#plt.style.available

squares = [1, 4, 9, 16, 25]
values = [1,2,3,4,5]

plt.style.use('Solarize_Light2')
fig, ax = plt.subplots()
ax.plot(values, squares, linewidth = 3)

# set chart title and label axes
ax.set_title('Square numbers', fontsize = 16, loc = 'center')
ax.set_xlabel('Values', fontsize = 14)
ax.set_ylabel('Square of values', fontsize = 14)

# set sizzze of tick labels
ax.tick_params(labelsize = 14)

plt.show()
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.scatter(x = 4, y = 2, s = 200)


ax.set_title('Value')
ax.set_label('x value')
ax.set_ylabel('y value')
plt.show()
```

```python
x_values = [3,8,10]
y_values = [1,4,12]

fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s =200)
plt.show()
```

```python
x_values = list(range(1,10))
y_values = [x**2 for x in x_values]

fig, ax = plt.subplots()
ax.scatter(x_values, y_values,c = y_values, cmap = 'Accent')
plt.show()
```

15-1. Cubes: A number raised to the third power is a cube. Plot the first five
cubic numbers, and then plot the first 5,000 cubic numbers.

```python
def plot_line(x, y, title, vertical_label, horizontal_label ):
    ax.plot(x, y, linewidth = 3)
    ax.set_title(title)
    ax.set_xlabel(vertical_label)
    ax.set_ylabel(horizontal_label)
    
    
    
```

```python
cube_values = list(range(1,6))
cube = [x**3 for x in cube_values]

fig, ax = plt.subplots()
plot_line(cube_values, cube, 'Cube', vertical_label = 'values', horizontal_label= 'cube')
ax.tick_params(colors = 'red')
plt.show()
```

```python
import seaborn as sns


cube_values = list(range(1,6))
cube = [x**3 for x in cube_values]

sns.lineplot(x = cube_values, y = cube )
plt.show()
```
