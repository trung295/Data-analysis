```python
import requests
import json
from pathlib import Path

url = r'https://hacker-news.firebaseio.com/v0/topstories.json'
r = requests.get(url)

response_dict = r.json()
top_10_stories = response_dict[:10].copy()
submission_links = [f'https://hacker-news.firebaseio.com/v0/item/{submission_id}.json' for submission_id in top_10_stories]
stories_dicts =[]
for submission_link in submission_links:
    r = requests.get(submission_link)
    stories_dict = r.json()
    stories_dicts.append(stories_dict)
stories_dicts

```
