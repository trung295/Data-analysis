```python
"""Tasks:
1. Data Loading & Cleaning
Load the dataset into a Pandas DataFrame.
Check for missing values and handle them appropriately.
Convert the Date column into the correct datetime format.
Remove duplicates if any exist.
2. Data Exploration
Display basic statistics (mean, median, min, max, etc.).
Find the top 5 best-selling products.
Identify the category that generates the most revenue.
Find the city with the highest number of sales.
3. Data Visualization
Plot total sales over time.
Create a bar chart showing sales per product category.
Plot a histogram of customer ages.
Create a correlation heatmap to analyze relationships between numerical features.
4. Sales Trends & Insights
Identify the month with the highest sales.
Analyze the impact of discounts on sales.
Compare sales on weekdays vs. weekends.
Determine if there's a trend in sales based on customer age groups.
5. Reporting & Conclusion
Summarize key insights from the analysis.
Suggest recommendations to improve sales based on the findings."""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\Hoang\Data analysis\Sale data analysis\data\sample_sales_data.csv")
pd.set_option("display.max_columns", None)  # Show all columns
pd.set_option("display.width",1000) # increase column width
print(df.info())
print(df.head())


#Clean data
if df.isnull().sum().sum() >0:
    df =df.dropna()
if df.duplicated().sum()>0:
    df = df.drop_duplicates()

#Convert date to date format

df["Date"] = pd.to_datetime(df["Date"])

# Find the top 5 best-selling products.
top_products = df.groupby("Product")["Total Sales"].sum().reset_index()
top_products = top_products.sort_values("Total Sales",ascending=False).head()
print(f"Top 5 best selling products: \n{top_products}")

#Identify the category that generates the most revenue.

categories = df.groupby("Category")["Total Sales"].sum().reset_index()
categories = categories.sort_values("Total Sales",ascending=False).head(1)
print(f"Most revenue category is {categories.iloc[0,0]} with revenue {categories.iloc[0,1]}")

#Find the city with the highest number of sales.

cities = df.groupby("City")["Total Sales"].sum().reset_index()
cities = cities.sort_values("Total Sales",ascending =False).head(1)
print(f"City {cities.iloc[0,0]} has the highest sales {cities.iloc[0,1]}")

#Plot total sales over time.

df["Year-Month"] = df["Date"].dt.to_period("M")
months = df.groupby("Year-Month")["Total Sales"].sum().reset_index()
months["Year-Month"] = months["Year-Month"].astype(str)
x_months =months["Year-Month"]
y_Sales = months["Total Sales"]
plt.figure(figsize=(10, 5))
plt.plot(x_months,y_Sales,marker="o",label="Total sale")


plt.xlabel("Total Sale($)")
plt.ylabel("Month")
plt.title("Total sale over time")
plt.grid(axis="y", linestyle="--")
plt.legend()
plt.xticks(rotation =45)
plt.savefig("sales_over_time_line.png", dpi=300, bbox_inches="tight")
plt.show()

#Create a bar chart showing sales per product category.

categories = df.groupby("Category")["Total Sales"].sum().reset_index()
categories = categories.sort_values("Total Sales",ascending=False)
x_category = categories["Category"]
y_value = categories["Total Sales"]
plt.figure(figsize=(10, 5))
plt.bar(x_category,y_value,color=["red","blue","green","yellow","purple"])

# Labels & title
plt.xlabel("Product Category")
plt.ylabel("Total Sale($)")
plt.title("Total Sales Per Product")
plt.grid(axis="y",linestyle="--")
plt.xticks(rotation=45)
plt.savefig("sales_per_product_chart.png", dpi=300, bbox_inches="tight")
plt.show()

#Plot a histogram of customer ages.

ages = df["Customer Age"]
hist, bin_edges = np.histogram(ages, bins='auto')
plt.figure(figsize=(10, 5))
plt.hist(ages, bins=bin_edges, color="skyblue", edgecolor="black")

# Labels & title
plt.xlabel("Customer Age")
plt.ylabel("Frequency")
plt.title("Distribution of Customer Ages")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show plot
plt.savefig("customer_ages_histogram.png", dpi=300, bbox_inches="tight")
plt.show()

#Create a correlation heatmap to analyze relationships between numerical features.

df_numeric  = df.drop(columns = ["Date","Product","Category","City","Year-Month"])
matrix = df_numeric.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(matrix, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")

# Show plot
plt.savefig("Numerical_correlation_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()    

#Identify the month with the highest sales.

months_highest = months.sort_values("Total Sales",ascending=False).head(1)
months_highest["Month"] = months_highest["Year-Month"].str[-2:]
months_highest["Year"] = months_highest["Year-Month"].str[:4]
print(f"Month {months_highest.iloc[0,2]} year {months_highest.iloc[0,3]} is the month with the highest sales {months_highest.iloc[0,1]}")

# Analyze the impact of discounts on sales.

df_sale_discount = df[["Total Sales","Discount"]]
df_sale_discount.corr()

plt.figure(figsize=(10,6))
sns.heatmap(df_sale_discount.corr(), annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title("Correlation Heatmap between sales and discount")
plt.show()

sns.scatterplot(x=df["Discount"], y=df["Total Sales"],alpha=0.5, color="blue", edgecolor="black")
plt.xlabel("Discount (%)")
plt.ylabel("Total Sales ($)")
plt.title("Impact of Discount on Sales")
plt.grid(True)
plt.savefig("sale_discount_scatter.png", dpi=300, bbox_inches="tight")
plt.show()

#Compare sales on weekdays vs. weekends.

df_weeks = df[["Date","Total Sales"]].copy()
df_weeks["Days"] = df_weeks["Date"].dt.day_name()
df_weeks["Days"] = df_weeks["Days"].apply(lambda x: 
                                          "Weekend" if x =="Saturday" or x == "Sunday" else
                                          "Weekday")
df_weeks_sale = df_weeks.groupby("Days")["Total Sales"].sum().reset_index()

x_week = df_weeks_sale["Days"]
y_week_sale = df_weeks_sale["Total Sales"]

plt.figure(figsize=(10,6))
plt.bar(x_week,y_week_sale,color=["red","blue"])
plt.ylabel("Total Sales($)")
plt.title("Comparision between weekdays and weekend")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle = "--")

for i, value in enumerate(df_weeks_sale["Total Sales"]):
    plt.text(i, value + 500, f"${value:,.0f}", ha="center", fontsize=12, fontweight="bold")
plt.savefig("sale_weekday_chart.png", dpi=300, bbox_inches="tight")
plt.show()

#Determine if there's a trend in sales based on customer age groups.

df_age_group = df[["Customer Age","Total Sales"]].copy()
df_age_group["Age group"] = df_age_group["Customer Age"].apply(lambda x:
                                                               "Teenage" if x<=15 else
                                                               "Young" if x<=21 else
                                                               "Aldult" if x<=45 else
                                                               "Senior" if x<=60 else
                                                               "Old")
age_group = df_age_group.groupby("Age group")["Total Sales"].sum().reset_index()
age_group = age_group.sort_values("Total Sales")


plt.figure(figsize=(10,6))
plt.bar(age_group["Age group"],age_group["Total Sales"], color =["Red","Blue","Yellow","Green","Purple"])

# Labels & title
plt.xlabel("Age Groups")
plt.ylabel("Total Sale($)")
plt.title("Total Sales base on customer age group")
plt.grid(axis="y",linestyle="--")
plt.xticks(rotation=45)

for i, value in enumerate(age_group["Total Sales"]):
    plt.text(i, value + 500, f"${value:,.0f}", ha="center", fontsize=12, fontweight="bold")

plt.savefig("customer_age_chart.png", dpi=300, bbox_inches="tight")
plt.show()







```
